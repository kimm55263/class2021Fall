```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
# parameter setting


amp = 1          # range [0.0, 1.0]
# 아래 위로 증폭; 시간+사인 곡선을 만들 텐데 amp가 1이다 0~1이다 = -1~1까지 가는 것; 2이면 -2~2이다

sr = 10000       # sampling rate, Hz

dur = 0.5        # in seconds 
# 끝날 때의 sample은 5000번째 샘플일 것

freq = 100.0     # sine frequency, Hz
# 태극문양이 1초에 총 100개가 들어가있다
# 0.5초 안에 50번 반복될 것

phase = np.pi/2  # phase, radian
# 90도에서 시작을 할 것이다

# OVERALL EXPLANATION
# 어떤 사인 곡선이 있는데, 1초에 몇 개의 샘플이 들어있는가? 태극문양 하나 1초에 하나 freq 1HZ
# sample이 10개가 있다 그러면 점이 10개 있는 것
# 샘플링r이 1초에 얼마나 자세한 정보를 전달하는가? resolution, sr을 넘어서는 frequency는 표현할 수 없음
# sr 반에 해당되는 것이 maximum frequency. sr이 44100 - 이것의 반이 해당 sr이 표현할 수 있는 최대한 frequency임
# 시간과 연결을 시켜야 함, 필수 요소는 sr임, radian의 정보를 sr의 정보로 바꾸어 줄 수 있음
# sr이 10HZ이다 그러면 하나의 sample이 의미하는 것은 1/10초
# sr = 10000HZ이다 그러면 하나의 sample이 담당하는 것은 1/10000초
```

```python
# generate samples (인덱스)
nSamp = sr * dur
samples = np.arange(1, nSamp+1)

# 입력은 각도값인 것 출력은 사인의 value값
# 여기에는 시간이 없음, 그래서 시간을 만들어 주어야 함
# 먼저, sr * duration을 하면 총 샘플이 5000개 있다 = nSamp에 들어가 있음
# 샘플에 번호를 만들어 주는 것 1에서 5000까지 숫자를 만들어 주는 것
```

```python
# convert samples to time
t = samples/sr

# 샘플 인덱스에 sr로 나누어 주면 1/10000 ~ 50000/100000까지 있는 것 - 이게 시간
# t라는 리스크 안에 있는 원소들을 각각 시간으로 나타내는 것
# 만약 sr=10000, dur=3일 경우 nSamp = 30000이 되고
# t는 1/10000초 2/10000초~ 9999/10000초...30000/10000초를 0.00001초부터 시작해서 3초까지 10000개의 공간을 만드는 것
# 여기는 5000개의 time tips들이 생김
```

```python
# convert time to angle in radian
theta0 = t * 2*np.pi * freq

# 0초에서 1초까지 시간이 있다고 할 때, fre1이다 - 그러면 한 바퀴도는데 2파이가 됨
# 이것이 0에서 2pi로 바뀌게 됨
```

```python
# shift angle by phase
theta = theta0 + phase

# Phase를 shift해주는 것
```

```python
# generate signal by sin-phasor
s = amp * np.sin(theta)
```

```python
# plot the entire signal
plt.plot(t, s, ':')
plt.xlabel('time (s)')
plt.ylabel('real')
```

```python
# plot a certain of signals, 확대해서 보고싶다
sampleRng = np.arange(0,100) # 100번째까지 볼 것이다
# 시험 문제는 100개의 샘플이 있을 때, frequency가 몇 개인지 계산할 수 있어야 함

plt.plot(t[sampleRng], s[sampleRng], ':')
plt.xlabel('time (s)')
plt.ylabel('real')
```

```python
ipd.Audio(s, rate=10000) # sr의 정보를 꼭 적어주어야 함
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta

  # 시간이 있는 사인 곡선을 만들어 낸다. 입력과 출력의 관계로 function을 만들어낼 수 있어야 한다
```
